# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# generate synthetic data
# Assume we want 5 features and 1000 samples
num_samples = 1000
mean = [0,0,0,0,0]   # mean of each feature
cov_matrix = [[1.0, 0.8, 0.5, 0.3, 0.1],
              [0.8, 1.0, 0.6, 0.2, 0.3],  # Covariance matrix
              [0.5, 0.6, 1.0, 0.4, 0.2],
              [0.3, 0.2, 0.1, 0.7, 1.0],
              [0.1, 0.3, 0.2, 0.7, 1.0]]

# Generate multivariate normal data with the specified mean and covariance
data = np.random.multivariate_normal(mean, cov_matrix, size=num_samples)

df = pd.DataFrame(data, columns=['Feature1','Feature2','Feature3','Feature4','Feature5'])

# compute the correlation matrix 
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,cmap='coolwarm', annot = True)
plt.title('Correlation Matrix heatmap')
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(30)

# Generate synthetic data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create the 3D scatter plot
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=df['X'], 
    y=df['Y'], 
    z=df['Z'], 
    mode='markers',
    marker=dict(
        size=6,  # Size of the marker
        color=df['Z'],  # Color the marker based on 'Z' values
        colorscale='Viridis',  # Color scale for the markers
        opacity=0.8
    )
))

# Update the layout
fig.update_layout(scene=dict(
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    zaxis_title='Z Axis'),
    title="3D Scatter Plot"
)

# Show the plot
fig.show()
```

```python
np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title('Heatmap of Sales Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

# 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)
```

# 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 50  # Marker sizes
colors = np.random.choice(["red", "green", "blue", "orange", "purple"], 100)  # Marker colors

# Prepare the data for Bokeh
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot with title and axis labels
p = figure(title="Scatter Plot with Random Data", x_axis_label="X", y_axis_label="Y", width=700, height=500)

# Add scatter markers
p.scatter(x='x', y='y', size='sizes', color='colors', fill_alpha=0.6, source=source)

show(p)
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
        fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
        counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="", 
           y_axis_label="Counts", height=350)  # Changed plot_height to height

# Add vertical bars to the plot
p.vbar(x='fruits', top='counts', width=0.6, source=source, 
       fill_color=factor_cmap('fruits', palette=["#c9d9d3", "#718dbf", "#e84d60", "#ddb7b1"], factors=fruits))

# Remove grid lines and set y-axis start
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

# 4.Create a Bokeh histogram to visualize the distribution of the given data.
    data_hist = np.random.randn(1000)
    hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

output_notebook()

# Create a figure for the histogram
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Count', 
           height=400, 
           width=700)

# Add the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')

# Show the plot
show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.
        data_heatmap = np.random.rand(10, 10)
        x = np.linspace(0, 1, 10)
        y = np.linspace(0, 1, 10)
        xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

output_notebook()

# Create a figure for the heatmap
p = figure(title="Heatmap of Random Data",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           tools='')

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data['value'].min(), high=data['value'].max())

# Add rectangles for the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=data, line_color=None, fill_color=mapper)

# Create a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
